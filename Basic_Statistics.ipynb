##Basic Statistics

```python
import numpy as np
import pandas as pd

data=[12,45,2,3,7,4,5,7,8,9,12,34,56,78,3,25,10,97,61]
```

```python
mean=np.mean(data)
median=np.median(data)
mode=pd.Series(data).mode()[0]
variance=np.var(data)
std_dev=np.std(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

##Handling missing data

```python
data=pd.Series([10,20,30,np.nan,40,50])
missing_values=data.isna().sum()
print(missing_values)
```

```python
#mean imputation
mean_imputed=data.fillna(data.mean())
print(mean_imputed)
```

```python
#median imputation
median_imputed=data.fillna(data.median())
print(median_imputed)
```

```python
#mode imputation
mode_imputed=data.fillna(data.mode())
print(mode_imputed)
```

##Visualization Basics

```python
import matplotlib.pyplot as plt
import seaborn as sns

months=['Jan','Feb','Mar','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
sales=[500,453,456,897,236,800,457,906,210,300,670,900]

#Bar chart
plt.bar(months,sales)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.show()
```

```python
#Line chart
plt.plot(months,sales)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.show()
```

```python
#Pie chart
plt.figure(figsize=(6,6))
plt.pie(sales,labels=months,autopct='%1.1f%%')
plt.title('Monthly Sales')
plt.show()
```

```python
#scatter plot
plt.scatter(months,sales)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.show()
```

```python
#pair plot
df=pd.DataFrame({'months':months,'sales':sales})
sns.pairplot(df)
```
